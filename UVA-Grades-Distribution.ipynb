```python
# This is a prototype database that pulls grade distributions from all classes
# ending in Spring of 2023. You need one file that will be included in the
# GitHub folfer. After, just enter a course and its number!
# It then shows a graph of their distributions.
# I hope to one day make the data pulling easier with an api maybe? Maybe I can
# turn it into a website too. I would also need to adjust for case sensitivity
# and error handle if there is a wrong input.
# I also need to improve the looks of the graphs in the future.

# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Loading Spring 2023 Data
grades = pd.read_csv("spring-2023-grades.csv")

# Input Prompts
subject_selection = input("Enter Course Subject Ex) ECON - ")
number_selection = int(input("Enter Course Number Ex) 3010 - "))

# Filtering
grades = grades.loc[grades["Subject"] == subject_selection]
grades = grades.loc[grades["Catalog Number"] == number_selection]

# Preparing Data Frame to Show the Grade Letters
grades = grades.iloc[:, 11:20]

# Calculating Data and Filtering
count = grades.sum()

# Count Distribution
count.plot(kind = "bar", width = .8, color = "orange")
plt.xlabel("Grade")
plt.ylabel("# of Students")
plt.title(subject_selection + " " + str(number_selection) + " Grade Distribution (Spring 2023)")

plt.show()

# Letter Grade Breakdown by Percentage

plt.pie(count, labels = grades.columns, autopct = "%1.1f%%")
plt.title(subject_selection + " " + str(number_selection) + " Grade Breakdown by %")
plt.show()
```
